```python
#importar tabela

import pandas as pd

tabela = pd.read_csv("clientes.csv")

tabela = tabela.drop(columns="id_cliente")
display(tabela)
display(tabela.info())
```

```python
#aplicar codificador em todas as colunas
import sklearn.preprocessing

codificador = sklearn.preprocessing.LabelEncoder()

for coluna in tabela.columns:
    if tabela[coluna].dtype == "object" and coluna != "score_credito":
        tabela[coluna] = codificador.fit_transform(tabela[coluna])

display(tabela.info())
```

```python
#Importar a IA

x = tabela.drop(["score_credito"], axis=1)
y = tabela["score_credito"]

from sklearn.model_selection import train_test_split

x_treino,x_teste, y_treino,y_teste = train_test_split(x,y, test_size=0.3)
```

```python
#Criar a IA

from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

modelo_arvore = RandomForestClassifier()
modelo_Knn = KNeighborsClassifier()
```

```python
#Treinar a IA

# Criar as instâncias dos modelos
modelo_arvore = RandomForestClassifier()
modelo_Knn = KNeighborsClassifier()

# Treinar os modelos
modelo_arvore.fit(x_treino, y_treino)
modelo_Knn.fit(x_treino, y_treino)
```

```python
#Avaliar modelo

from sklearn.metrics import accuracy_score

previsao_arvore = modelo_arvore.predict(x_teste)
previsao_knn = modelo_Knn.predict(x_teste)

display(accuracy_score(y_teste,previsao_arvore))
display(accuracy_score(y_teste, previsao_knn))
```

```python
#Fazendo novas previsões com minha IA

novos_clientes = pd.read_csv("novos_clientes.csv")
display(novos_clientes)

for coluna in novos_clientes.columns:
    if novos_clientes[coluna].dtype == "object" and coluna != "score_credito":
        novos_clientes[coluna] = codificador.fit_transform(novos_clientes[coluna])

previsoes = modelo_arvore.predict(novos_clientes)
print(previsoes)
```
